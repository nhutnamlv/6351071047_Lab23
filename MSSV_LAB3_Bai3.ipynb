```python
print("Trần Nhựt Nam_6351071047 - a) Đọc dữ liệu MNIST")

from keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print("Trần Nhựt Nam_6351071047 - b) Kích thước dữ liệu")

print("Train images:", train_images.shape)
print("Train labels:", train_labels.shape)
print("Test images:", test_images.shape)
print("Test labels:", test_labels.shape)
```

```python
print("Trần Nhựt Nam_6351071047 - c) Chuẩn hóa dữ liệu")

train_images = train_images / 255.0
test_images  = test_images  / 255.0
```

```python
print("Trần Nhựt Nam_6351071047 - d) Tạo hàm create_model")

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten

def create_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(10, activation='softmax')
    ])

    model.compile(
        optimizer='adam',
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
print("Trần Nhựt Nam_6351071047 - e) Huấn luyện K-Fold")

from sklearn.model_selection import KFold
import numpy as np

kfold = KFold(n_splits=10, shuffle=True, random_state=42)

accuracy_per_fold = []
loss_per_fold = []
models = []

fold_no = 1
for train_idx, val_idx in kfold.split(train_images):

    print(f"\n========== FOLD {fold_no} ==========")

    X_train_fold = train_images[train_idx]
    y_train_fold = train_labels[train_idx]
    X_val_fold   = train_images[val_idx]
    y_val_fold   = train_labels[val_idx]

    model = create_model()

    history = model.fit(
        X_train_fold, y_train_fold,
        epochs=5,
        batch_size=32,
        validation_data=(X_val_fold, y_val_fold),
        verbose=0
    )

    models.append(model)

    loss, acc = model.evaluate(X_val_fold, y_val_fold, verbose=0)
    loss_per_fold.append(loss)
    accuracy_per_fold.append(acc)

    print(f"Fold {fold_no} - Loss: {loss} - Accuracy: {acc}")

    fold_no += 1
```

```python
print("Trần Nhựt Nam_6351071047 - f) Mean & Std")

print("Mean Accuracy:", np.mean(accuracy_per_fold))
print("Std Accuracy :", np.std(accuracy_per_fold))
print("Mean Loss    :", np.mean(loss_per_fold))
print("Std Loss     :", np.std(loss_per_fold))
```

```python
print("Trần Nhựt Nam_6351071047 - g) Vẽ biểu đồ")

import matplotlib.pyplot as plt

plt.plot(accuracy_per_fold)
plt.title("Accuracy per fold")
plt.show()

plt.plot(loss_per_fold)
plt.title("Loss per fold")
plt.show()
```
