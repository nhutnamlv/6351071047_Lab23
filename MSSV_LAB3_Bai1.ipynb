```python
print("Trần Nhựt Nam_6351071047 - Đọc dữ liệu Iris")

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

iris = load_iris()
X = iris.data
y = iris.target
```

```python
print("Trần Nhựt Nam_6351071047 - Khảo sát dữ liệu")

print("Tên các thuộc tính:", iris.feature_names)
print("Tên các lớp:", iris.target_names)

print("Kích thước X:", X.shape)
print("Kích thước y:", y.shape)

print("5 mẫu dữ liệu đầu tiên:")
print(X[:5])
print("5 nhãn đầu tiên:", y[:5])

df = pd.DataFrame(X, columns=iris.feature_names)
print(df.describe())
```

```python
print("Trần Nhựt Nam_6351071047 - Khảo sát dữ liệu")

print("Tên các thuộc tính:", iris.feature_names)
print("Tên các lớp:", iris.target_names)

print("Kích thước X:", X.shape)
print("Kích thước y:", y.shape)

print("5 mẫu dữ liệu đầu tiên:")
print(X[:5])
print("5 nhãn đầu tiên:", y[:5])

df = pd.DataFrame(X, columns=iris.feature_names)
print(df.describe())
```

```python
print("Trần Nhựt Nam_6351071047 - Chuẩn hóa dữ liệu")

X_norm = X / X.max(axis=0)
```

```python
print("Trần Nhựt Nam_6351071047 - Chia dữ liệu 80/20")

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y, test_size=0.2, random_state=42
)
```

```python
print("Trần Nhựt Nam_6351071047 - Decision Tree")

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

dt = DecisionTreeClassifier(random_state=0)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

print("Accuracy Decision Tree:", accuracy_score(y_test, y_pred_dt))
print(classification_report(y_test, y_pred_dt))

cm = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm, annot=True)
plt.title("Decision Tree - Confusion Matrix")
plt.show()
```

```python
print("Trần Nhựt Nam_6351071047 - Naive Bayes")

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)

print("Accuracy Naive Bayes:", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))

cm = confusion_matrix(y_test, y_pred_nb)
sns.heatmap(cm, annot=True)
plt.title("Naive Bayes - Confusion Matrix")
plt.show()
```

```python
print("Trần Nhựt Nam_6351071047 - KNN")

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

print("Accuracy KNN:", accuracy_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))

cm = confusion_matrix(y_test, y_pred_knn)
sns.heatmap(cm, annot=True)
plt.title("KNN - Confusion Matrix")
plt.show()
```

```python
print("Trần Nhựt Nam_6351071047 - Neural Network")

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(10, activation='relu', input_shape=(4,)),
    Dense(20, activation='relu'),
    Dense(3, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=1)

y_pred_nn = np.argmax(model.predict(X_test), axis=1)

print("Accuracy NN:", accuracy_score(y_test, y_pred_nn))
print(classification_report(y_test, y_pred_nn))

cm = confusion_matrix(y_test, y_pred_nn)
sns.heatmap(cm, annot=True)
plt.title("Neural Network - Confusion Matrix")
plt.show()
```
